```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
```

```python
options = Options()
options.add_argument("--start-maximized")

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=options
)

url = "https://www.goodreads.com/search?q=python+programming"
driver.get(url)

time.sleep(5)
```

```python
books = driver.find_elements(By.XPATH, "//tr[@itemtype='http://schema.org/Book']")

print("Total books found:", len(books))

data = []

for book in books:

    try:
        title = book.find_element(By.XPATH, ".//a[@class='bookTitle']").text
    except:
        title = "N/A"

    try:
        author = book.find_element(By.XPATH, ".//a[@class='authorName']").text
    except:
        author = "N/A"

    try:
        rating_text = book.find_element(By.XPATH, ".//span[@class='minirating']").text
        rating = rating_text.split("—")[0].strip()
    except:
        rating = "N/A"

    try:
        published = rating_text.split("published")[-1].strip()
    except:
        published = "N/A"

    try:
        link = book.find_element(By.XPATH, ".//a[@class='bookTitle']").get_attribute("href")
    except:
        link = "N/A"

    data.append({
        "Title": title,
        "Author": author,
        "Rating": rating,
        "Published Year": published,
        "Book Link": link
    })

df = pd.DataFrame(data)

print("Total data rows:", len(df))
df
```

```python
df.to_csv("goodreads_books_data.csv", index=False, encoding="utf-8-sig")
print("CSV saved successfully")
```

```python
driver.quit()
```
