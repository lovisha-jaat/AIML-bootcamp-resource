 - Retrieving web page content: Server responds with the web page content, usually HTML format

```python
import requests
```

```python
url = "https://realpython.github.io/fake-jobs/?"
response = requests.get(url)
response.status_code
```

```python
html_code = response.text
html_code
```

```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
code = BeautifulSoup(html_code,'html.parser')
code
```

```python
print(code.prettify())
```

    🌐 Step 4 : Extracting Data 🌐
    
</div>

```python
details = code.find_all('div',class_= 'card-content')
details
```

```python
data = []
for d in details:
    title = d.find('h2',class_='title is-5')
    subtitle = d.find('h3',class_= 'subtitle is-6 company')
    location = d.find('p',class_= 'location')
    time = d.find('p',class_= 'is-small has-text-grey')
    title_el = title.get_text(strip = True) if title else ""
    subtitle_el = subtitle.get_text(strip = True) if subtitle else ""
    location_el = location.get_text(strip = True) if location else ""
    time_el = time.get_text(strip = True) if time else ""
    data.append({
        'Job_name' : title_el,
        'company_name' : subtitle_el,
        'location' : location_el,
        'time' : time_el
    })
data
```

```python
df = pd.DataFrame(data)
df
```
